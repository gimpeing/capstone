## 6.1 Metrics

To evaluate the effectiveness of the classification model, metrics used to assess are:

1. **$$Precision = \frac{TP}{FP+TP}$$**

**Precision** provides the *positive predictive value, the proportion of samples that belong in category $x$ that are correctly placed in category $x$*. Example, among all predicted murmur heartbeat, how many did I predict correctly? High precision is with low FP (predict to be murmur but actually is not).

2. **$$f1 = 2\frac{Precision*Recall}{Precision + Recall}$$**

**Weighted average for precision and recall**. This score takes both FN and FP into account as $$Recall = \frac{TP}{FN+TP}$$ 

This is better metric for evaluating an uneven class distribution problem, which is the case in this project. 


3. $$Accuracy = \frac{TP+TN}{All predicts}$$

### Kaggle Evaluation Metrics

#### 1. Precision

#### 2. Youden's Index
Youden's index has traditionally been used to compare diagnostic abilities of two test, by **evaluating the algorithm's ability to avoid failure**. 
Dataset A: evaluate the Youden's Index of the Artifact category
Dataset B: evaluate the Youden's Index of problematic heartbeats (Murmur and Extrasystole categories combined)

$$\gamma = sensitivity - (1 - specificity)$$

where $$sensitivity = \frac{TP}{TP+FN}$$ , $$specificity = \frac{TN}{FP+TN}$$

#### 3. F-score
Dataset A only: set $\beta$ = 0.9 and evaluate the F-score of problematic heartbeats of **murmur and extra heart sound categories combined**.
$$F-score = \frac{(\beta^2+1)*precision*sensitivity}{\beta^2*precision+sensitivity}$$

#### 4. Discriminant Power
Dataset B only: evaluate how well an algorithm distinguishes between positive and negative examples. The algorithm is a poor discriminant if DP<1, limited if DP<2, fair if DP<3, and good in other cases. Here we calculate the DP of problematic heartbeats (**murmur and extrasystole categories combined**)
$$DP = \frac{\sqrt{3}}{\pi}(logX + logY)$$

where $$X= \frac{sensitivity}{1-sensitivity}$$ and
$$Y=\frac{specificity}{1-specificity}$$

# 7 Conclusions and Recommendation

The **goal** of this project is to use machine learning methods to **classify heartbeat sounds** from audio collected from stethoscopes into normal versus various non-normal heartbeat sounds. The heartbeat sounds' data is collected in real-world situations and some of them contains background noise. The differences between heart sounds correspoding to different heart symptoms sometimes extremely subtle and is challenging in separating them. Success in classifying this form of data requires extremely robust classifiers.

From the modelling results analysis, **CWT feature extraction techniques and using 2D CNN** to classify the heartbeat audio categories stand out to be better than MFCC feature extraction + 2D CNN and CWT + XGBOOST. It has higher precision, and f1 scores, as well as it is able to generalize better on unseen data, as it consistently perform well in unseen TEST dataset, and unseen unlabel dataset (for both set_a & set_b). In addition, it perform better compared to a reference from a technical paper using the SAME datasets (unlabel set_a nad set_b).

However, the model's efficiency in classifying normal and various non-normal heartbeat sounds still need to further improve as the precision, f1 scores for some of the heatbeat category is still low. 

#### Recommendation for the improvements:
1. More data. The label datasets are only 585 and 60% of them are from normal heartbeat category. Thus, more data for both normal and non-normal heartbeat categories are needed to train the clasiffier model. Audio wav with at least 3 sec is desirable as some of the audio wav is as short as 0.7sec. 

2. Segmentation approach in preprocessing could be improved. This project use a segment of T=3sec for training and validation, where post-zero padding is apply on audio shorter than 3sec. Segmentation could be improve to first detect the position of heartbeat S1 & S2, and segment the audio into smaller chunk by locating the begining of the heartbeat and ensuring the segmented audio consist of at least x amount of heart cycles/ hearbeat pairs. 

3. CWT feature extraction optimization by identifying the least number of CWT coefficients that does not loose out the important details of the audio signal. This will reduce the computation rsources significantly in CNN.

4. Explore RNN (improve the model efficiency in classifying the heartbeat category) and XGBOOST (maintain the model efficiency but with much lesser computational time)
